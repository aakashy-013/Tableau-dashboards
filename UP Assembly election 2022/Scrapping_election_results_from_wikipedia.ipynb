## 1. Importing Packages

```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
```

```python
response = requests.get('https://en.wikipedia.org/wiki/Uttar_Pradesh_Legislative_Assembly')
```

```python
html = response.content
```

## 2. Scrapping data from wikipedia

```python
soup = BeautifulSoup(html,'lxml')
```

```python
soup
```

```python
table = soup.find_all('table')
len(table)
```

```python
table[3]
```

```python
header = table[3].find_all('th') 
header
```

```python
# storing header name in variable columns
columns = []
for i in header:
    columns.append(i.text.strip())
```

```python
columns
```

```python
# stroring district names in district columns
dist = table[3].find_all('td',align='center')
district=[]
for i in dist:
    district.append(i.text.strip())
print(len(district))
district
```

```python
# This line of code is to look into the pattern of storing data in table
t = table[3].find_all('td')
t
```

## 3. Creating DataFrame

```python
data = pd.DataFrame(columns = columns)
df = pd.DataFrame(columns = ['District'])
```

```python
k=0
for i in list:
    for j in range(int(len(i)/8)):
        data.loc[len(data.index)] = i[j*8:j*8+8]
        df.loc[len(data.index)] = district[k]
    k=k+1
```

```python
# here we removed those blank columns
data.drop(['u1','u2'],axis = 1,inplace=True)
```

```python
data
```

```python
df.reset_index(drop=True,inplace=True)
# added district column in dataframe after removing **District** from each district name
df['District'] = df['District'].str.replace('District','')
df
```

```python
# Here, data and df are concatenated
data = pd.concat([data,df['District']],ignore_index=False,sort=False,axis = 1)
```

## 4. Saving dataframe

```python
# dataframe is saved as District_wise_data.csv
data.to_csv('District_wise_data.csv',index=False)
```
